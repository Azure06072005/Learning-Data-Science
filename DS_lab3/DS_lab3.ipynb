### Pyplot API vs Object-Oriented API
* Pyplot APO --> Quickly
* OO API --> Advanced

#### Pyplot API

```python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x), color='blue', linestyle='dashed', label='sin(x)');
plt.plot(x, np.cos(x), color='r', linestyle='dashed', label='cos(x)')
plt.xlabel("Bien X")
plt.ylabel("sin(X)")
plt.title("Sin(x) and Cos(x)")
plt.axis("equal")
# plt.axis("tight")
# plt.axis([0, 4, -0.7, 0.6])
# plt.xlim(0, 4)
# plt.ylim(-0.75, 0.75)
plt.legend()
```

```python
#1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2. Setup plot
fig, ax = plt.subplots(figsize=(5, 5)) #Figure sizr = width & height fo the plot

#3. Plot the data
ax.plot(x, y)
ax.set(title="A Simple Plot", xlabel="x-axis", ylabel="y-axis");
```

## 2. Most common types of Matplotlib plots

#### Line

```python
x = np.linspace(0, 10, 100)
x[:5]
```

```python
## OO API plot a line chart 
fig, ax = plt.subplots()
ax.plot(x, x**3)
```

#### Scatter

```python
# Pyplot API
plt.scatter(x, np.exp(x));
```

```python
# OO API
# Prepare new data
nrg = np.random.RandomState(0)

x = nrg.randn(100)
y = nrg.randn(100)
colors = nrg.rand(100)
sizes = 1000 * nrg.rand(100)

fig, ax = plt.subplots()
img1 = ax.scatter(x, y, s=sizes, c = colors, cmap='viridis', alpha=0.5);

fig.colorbar(img1)
```

### 3. Bar
- Vertical
- Horizontal

```python
# Prepare data from a python dictionary
sort_drink_prices = {"Coke": 10,
                     "Pessi": 12,
                     "Sprite": 8}
fig, ax = plt.subplots()
ax.bar(sort_drink_prices.keys(), sort_drink_prices.values());
ax.set(title="Bach Hoa Xanh's sort drink prices", ylabel="Price (VND)")
```

```python
fig, ax = plt.subplots()
ax.barh(list(sort_drink_prices.keys()), list(sort_drink_prices.values()));
```

### 4. Histogram

```python
# Prepare data 
np.random.seed(42)
student_height = np.random.normal(170, 10, 250)
```

```python
student_height[:10]
```

```python
# Pyplot API
plt.hist(student_height, bins = 20);

```

### Subplots 
- Multi plots on the figure:

```python
# Option 1: 
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, 
                      ncols = 2, 
                      figsize = (10, 5))

# Plot data to each axis
ax1.plot(x, x/2)
ax1.plot(x, x**2, color='r')

ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(sort_drink_prices.keys(), sort_drink_prices.values())
ax4.hist(student_height, bins = 20)
```

```python
# Option 2: 
fig, ax = plt.subplots(nrows=2,
                       ncols=2, 
                       figsize=(10, 5))

# Plot data to each axis
ax[0, 0].plot(x, x/2);
ax[0, 0].plot(x, x**2);

ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(sort_drink_prices.keys(), sort_drink_prices.values());
ax[1, 1].hist(student_height, bins=20);
```
